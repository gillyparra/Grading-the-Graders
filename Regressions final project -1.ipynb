In this file we run the regressions with our cleaned data

```python
# Simple OLS
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

df = pd.read_csv("C:/Users/Guill/OneDrive/Documents/Alltogetherall.csv")

result = smf.ols(formula='Avg ~ ContractEnded', data=df).fit()

print(result.summary())
```

```python
# OLS controlling for enrolled students
result = smf.ols(formula='Avg ~ ContractEnded+Enrolled', data=df).fit()

print(result.summary())
```

```python
# Regression with professor fixed effects
result = smf.ols(formula='Avg ~ ContractEnded + C(Professor)', data=df).fit()

print(result.summary())
```

```python
# Regression with professor fixed effects and course fixed effects
result = smf.ols(formula='Avg ~ ContractEnded + C(Course) + C(Professor)', data=df).fit()

print(result.summary())
```

```python
# Final regression with professor fixed effects and course fixed effects and students enrolled control. 
result = smf.ols(formula='Avg ~ ContractEnded + C(Course) + C(Professor)+ Enrolled', data=df).fit()

print(result.summary())
```
